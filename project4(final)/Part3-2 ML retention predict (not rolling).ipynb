```python
'''def dataset_split(df):
    train_Y = df['retain']
    print(train_Y.value_counts())

    drop_list =['MemberId', 'count', 'retain']
    train_X = df.drop(drop_list ,axis=1)
    column_name = train_X.columns.tolist()

    train_X = train_X.values
    train_X, test_X, train_Y, test_Y = train_test_split(train_X, train_Y, test_size=0.2)
    
    return train_X, test_X, train_Y, test_Y, column_name'''
    

def classification_model(df, model):
    print(datetime.now())
    
    #train test split
    train_Y = df['retain']
    print(train_Y.value_counts())

    drop_list =['MemberId', 'count', 'retain']
    train_X = df.drop(drop_list ,axis=1)
    column_name = train_X.columns.tolist()

    train_X = train_X.values
    train_X, test_X, train_Y, test_Y = train_test_split(train_X, train_Y, test_size=0.2)
    
    #model
    model = model
    model.fit(train_X, train_Y)
    print('model fit')
    
    ##cross validation score
    score = cross_val_score(model, train_X, train_Y, cv=5).mean()
    print("cross validation score: %0.2f (+/- %0.2f)" % (score.mean(), score.std() * 2))
    
    ##predict
    pred_Y = model.predict(test_X)
    probs_Y = model.predict_proba(test_X)
    probs_Y = probs_Y[:, 1] ## keep probabilities for the positive outcome only
    
    ##evaluation
    accy = accuracy_score(test_Y, pred_Y)
    f1 = f1_score(test_Y, pred_Y)
    recall = recall_score(test_Y, pred_Y)
    precision =precision_score(test_Y, pred_Y)
    auc = roc_auc_score(test_Y, probs_Y)
    ap = average_precision_score(test_Y, probs_Y) 
    print('evaluation without RFECV: accy: %0.3f, f1: %0.3f, recall: %0.3f, precision: %0.3f, auc: %0.3f, ap: %0.3f' % (accy, f1, recall, precision, auc, ap))
    
    ##confusion matrix
    labels = np.unique(test_Y)
    conf_m = confusion_matrix(test_Y, pred_Y, labels=labels)
    conf_m = pd.DataFrame(conf_m, index=labels, columns=labels)
    #prctg = conf_m/conf_sum
    print('confusion matrix without RFECV:', conf_m)
    
    #feature importance
    try:
        feature_importances = pd.Series(model.feature_importances_, index=column_name)
    except:
        feature_importances = pd.Series(['None']) 
    
    #rfecv
    '''selector= RFECV(model, step=1, cv=StratifiedKFold(2), scoring='accuracy')
    selector.fit(train_X, train_Y)
    print("Optimal number of features : %d" % selector.n_features_)
    feature_ranking = sorted(zip(selector.ranking_,column_name))
    for i in feature_ranking:
        print(i)
        
    pred_Y_rfe = selector.predict(test_X)
    probs_Y_rfe = selector.predict_proba(test_X)
    probs_Y_rfe = probs_Y_rfe[:, 1] ## keep probabilities for the positive outcome only
        
    ##evaluation
    accy_rfe = accuracy_score(test_Y, pred_Y_rfe)
    #f1_rfe = f1_score(test_Y, pred_Y_rfe)
    recall_rfe = recall_score(test_Y, pred_Y_rfe)
    #precision_rfe =precision_score(test_Y, pred_Y_rfe)
    auc_rfe = roc_auc_score(test_Y, probs_Y_rfe)
    print('evaluation with RFECV: accy: %0.3f, recall: %0.3f, auc: %0.3f' % (accy_rfe, recall_rfe, auc_rfe))
    
    ##confusion matrix
    labels = np.unique(test_Y)
    conf_m_rfe = confusion_matrix(test_Y, pred_Y_rfe, labels=labels)
    conf_m_rfe = pd.DataFrame(conf_m_rfe, index=labels, columns=labels)
    #prctg_rfe = conf_m_rfe/conf_sum_rfe
    print('confusion matrix with RFECV:', conf_m_rfe)'''
    print(datetime.now())
    
    return pred_Y, probs_Y, feature_importances, accy, f1, recall, precision, auc, conf_m, ap
    
```

```python
df_lst = [df_1st, df_2nd, df_3rd]
model_lst = [LogisticRegression(), RandomForestClassifier(), XGBClassifier()]

accy_lst = []
f1_lst = []
recall_lst = []
precision_lst = []
auc_lst = []
#conf_m_lst = []
feature_lst = []
ap_lst = []

for df in df_lst:
    for model in model_lst:
        pred_Y, probs_Y, feature_importances, accy, f1, recall, precision, auc, conf_m, ap = classification_model(df, model)
        accy_lst.append(accy)
        f1_lst.append(f1)
        recall_lst.append(recall)
        precision_lst.append(precision)
        auc_lst.append(auc)
        ap_lst.append(ap)
        #conf_m_lst.append(conf_m)
        feature_lst.append(feature_importances.to_dict())
    
```

```python
eva = pd.DataFrame()
eva['accy'] = accy_lst
eva['f1'] = f1_lst
eva['recall'] = recall_lst
eva['precision'] = precision_lst
eva['auc'] = auc_lst
eva['ap'] = ap_lst

eva.index = ['LR1', 'RF1', 'XGB1', 'LR2', 'RF2', 'XGB2', 'LR3', 'RF3', 'XGB3']
```
