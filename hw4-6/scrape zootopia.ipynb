```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
def get_document(url):
    res = requests.get(url)
    res.encoding = 'utf-8'
    doc = bs(res.text,'lxml')  #該頁的全部網頁原始碼
    return doc

```

```python
url = 'https://www.imsdb.com/scripts/Zootopia.html'
doc = get_document(url)
```

```python
content = doc.select('td > pre')
for ele in content:
    text = ele.get_text()
print(text)
```

```python
file = open("zootopia.txt","w")
file.write(text)
file.close()
```
